```python
import pandas as pd
import numpy as np
import os
import sys
from tqdm import tqdm
```

```python
# merge all asvspoof 2019 protocol files
train_df = pd.read_csv('ASVspoof2019.LA.cm.train.trn.txt', sep=' ', header=None, names=['spid', 'utt', 'phy', 'attack', 'label'])
dev_df = pd.read_csv('ASVspoof2019.LA.cm.dev.trl.txt', sep=' ', header=None, names=['spid', 'utt', 'phy', 'attack', 'label'])
eval_df = pd.read_csv('ASVspoof2019.LA.cm.eval.trl.txt', sep=' ', header=None, names=['spid', 'utt', 'phy', 'attack', 'label'])

full_df = pd.concat([train_df, dev_df, eval_df], ignore_index=True)
full_df
```
